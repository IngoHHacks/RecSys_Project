```python
from sklearn.cluster import KMeans
import pickle
import numpy as np
```

```python
# Load from the pickle file
with open('/kaggle/input/user-features-rs-project/user_features_dic.pkl', 'rb') as pickle_file:
    user_features_dic = pickle.load(pickle_file)
```

```python
user_features_values = np.array(list(user_features_dic.values()))
```

```python
user_features_keys = np.array(list(user_features_dic.keys()))
```

```python
#Number of clusters
num_clusters = 100

# Perform K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(user_features_values)

# You can also access the cluster centers if needed
cluster_centers = kmeans.cluster_centers_
```

```python
clusters_ids = []

# Print the cluster assignments for each array
for i in range(num_clusters):
    print(f'Cluster {i+1}:')
    
    current_cluster_ids = []
    
    for j, label in enumerate(cluster_labels):
        if label == i:
            print(f'Array {j+1}')
            current_cluster_ids.append(user_features_keys[j])
            if len(current_cluster_ids) > 4:
                break
            
    if len(current_cluster_ids) > 1:
        clusters_ids.append(current_cluster_ids)
```
