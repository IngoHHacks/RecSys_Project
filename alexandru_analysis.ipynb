### Check if both datasets have the same items (games) in their records

```python
first_dataset = pd.read_csv(data_folder + 'games_detailed_info.csv')
second_dataset = pd.read_csv(data_folder + 'games.csv')
```

```python
first_dataset
```

```python
second_dataset
```

```python
first_id_to_name = {first_dataset['id'][i]: first_dataset['primary'][i] for i in range(len(first_dataset['id']))}
```

```python
second_id_to_name = {second_dataset['BGGId'][i]: second_dataset['Name'][i] for i in range(len(second_dataset['BGGId']))}
```

```python
hit = 0

for key, value in first_id_to_name.items():
    
    if key in second_id_to_name and value == second_id_to_name[key]:
        hit += 1
```
